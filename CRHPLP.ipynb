### 階層ごとに説明変数の平均をだして比較してみる
これだけでも何が違いを産んでるのか結構見えてくるはず

```python
group_1 = train_data.query("Target==1") #df.query("条件式")で条件に合うデータを取り出せる。
group_2 = train_data.query("Target==2")
group_3 = train_data.query("Target==3")
group_4 = train_data.query("Target==4")
```

```python
groups = [group_1,group_2,group_3,group_4]
i = 1
for group in groups:
    print("*"*30)
    print("Group" + str(i))
    i += 1
    print(group.mean())
```

```python
groups_mean = pd.DataFrame({"G1":group_1.mean(),
                            "G2":group_2.mean(),
                            "G3":group_3.mean(),
                            "G4":group_4.mean()})
groups_mean = groups_mean.T
groups_mean
```

```python
n = 1
for i in groups_mean.columns:
    a = groups_mean.loc[:,[i]]
    plt.figure(figsize=(60,20))
    plt.subplot(6,23,n)
    n +=1
    plt.hist(a)
```
